## Turn the Water Quality Data into a  gdf Using  Given BNG Coordinates 

```python
gdf_sampling_points = gpd.GeoDataFrame(
    df_water_final, 
    geometry=gpd.points_from_xy(df_water_final['northing'], df_water_final['easting']),
    crs="EPSG:27700" 
)
gdf_sampling_points.head()
```

## Merge All Together by Classifying Sampling Points Based on Local Authority Geometry

```python
# convert sampling points into a gdf with Point geometries
gdf_sampling_points['geometry'] = gdf_sampling_points.apply(
    lambda row: Point(row['easting'], row['northing']), axis=1
)

# convert to gdf and set CRS to British National Grid (EPSG:27700)
gdf_sampling_points = gpd.GeoDataFrame(gdf_sampling_points, geometry='geometry', crs="EPSG:27700")
gdf = gdf.to_crs("EPSG:27700")  

# perform spatial join to assign each sampling point to a Local Authority
gdf_classified = gpd.sjoin(gdf_sampling_points, gdf[['LAD22NM', 'geometry']], how="left", predicate="within")

# keep only relevant columns
gdf_classified = gdf_classified[['sample_date_time', 'northing', 'easting', 'LAD22NM', 'determinand_label',	'determinand_def',	'result',	'determinand_unit',	'sample_type',	'year']]

print(gdf_classified.head())
```

```python
df_land.columns
```

```python
# Rename
gdf_classified.rename(columns={"LAD22NM": "local_authority_name"}, inplace=True)

# Remove the column axis name
gdf_classified.columns.name = None

# merging
df_merged = pd.merge(gdf_classified, df_land, on=['local_authority_name', 'year'])
```

## Columns in the Merged Dataset 

```python
df_merged.head()
```

```python
df_merged.columns
```

```python
# Pivot the DataFrame to wide format
df_wide = df_merged.pivot_table(
    index=[
        'sample_date_time', 'northing', 'easting', 'local_authority_name',
        'sample_type', 'year', 'industry', 'offices', 'retail', 'storage_and_warehousing',
        'industry_and_commerce_total', 'landfill_and_waste_disposal',
        'minerals_and_mining', 'minerals_and_landfill_total',
        'institutional_and_communal_accommodations', 'residential',
        'residential_total', 'highways_and_road_transport', 'transport_other',
        'utilities', 'transport_and_utilities_total', 'agricultural_buildings',
        'agricultural_land', 'agriculture_total', 'grand_total'
    ],
    columns="determinand_label",
    values="result",
    aggfunc="first"
).reset_index()

# Flatten column hierarchy
df_wide.columns.name = None  

# Drop rows with missing parameter measurements
df = df_wide.dropna().copy()

# Convert 'sample_date_time' to datetime format
df['sample_date_time'] = pd.to_datetime(df['sample_date_time'])

# Handle numeric columns stored as object/string types
numeric_columns = [
    'industry_and_commerce_total', 'landfill_and_waste_disposal',
    'minerals_and_mining', 'minerals_and_landfill_total', 'residential',
    'residential_total', 'highways_and_road_transport', 'transport_other',
    'utilities', 'transport_and_utilities_total', 'agricultural_buildings',
    'agricultural_land', 'agriculture_total', 'grand_total'
]

for column in numeric_columns:
    df[column] = df[column].astype(str).str.strip()
    df[column] = df[column].str.replace(r'[^0-9.]', '', regex=True)
    df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
# Create combined total columns
df['industry_total'] = df['industry'] + df['storage_and_warehousing']
df['transport_total'] = df['highways_and_road_transport'] + df['transport_other']

# Calculate percentage features relative to 'grand_total'
df['industry_percentage'] = (df['industry_total'] / df['grand_total']) * 100
df['minerals_percentage'] = (df['minerals_and_mining'] / df['grand_total']) * 100
df['residential_percentage'] = (df['residential_total'] / df['grand_total']) * 100
df['transport_percentage'] = (df['transport_total'] / df['grand_total']) * 100
df['agriculture_percentage'] = (df['agriculture_total'] / df['grand_total']) * 100

# Check missing values
print("Missing values before fillna():")
print(df.isnull().sum())

# Handle remaining missing values
df.fillna(0, inplace=True)
```

```python
columns_to_drop = [
    'northing', 'easting', 'sample_type', 'offices', 'retail',
    'industry_and_commerce_total', 'landfill_and_waste_disposal',
    'minerals_and_landfill_total', 'institutional_and_communal_accommodations',
    'residential', 'utilities', 'transport_and_utilities_total',
    'agricultural_buildings', 'agricultural_land', 'industry',
    'storage_and_warehousing', 'minerals_and_mining', 'residential_total',
    'highways_and_road_transport', 'transport_other', 'agriculture_total',
    'industry_total', 'transport_total'
]

# Drop columns:
existing_cols_to_drop = [col for col in columns_to_drop if col in df.columns]

df = df.drop(existing_cols_to_drop, axis=1)
```
